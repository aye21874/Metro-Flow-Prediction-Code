```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(list2, test_size = 0.2)
x1 = train.iloc[:,0:45]
y1 = train.iloc[:,45]
x1_test = test.iloc[:,0:45]
y1_test = test.iloc[:,45]

from sklearn import neighbors
model = neighbors.KNeighborsRegressor(n_neighbors = 12,weights = 'uniform')
model.fit(x1,y1)

```

```python
from sklearn.metrics import mean_squared_error
predictions = model.predict(x1_test)
knn_mse = mean_squared_error(y1_test,predictions)
knn_rmse = np.sqrt(knn_mse)
knn_rmse
```
